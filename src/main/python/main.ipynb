```python
import socket
import json
```

```python

def handle_client_connection(client_socket):
    query = client_socket.recv(1024).decode('utf-8').strip()
    print("Nhận query từ client:", query)

    # Xử lí query ở đây và tạo kết quả
    response = query_handler.query(query)

    # Gửi kết quả về client dưới dạng JSON
    if response:
        client_socket.sendall(response.encode('utf-8'))

    client_socket.close()
```

```python
def main():
    host = 'localhost'
    port = 8000

    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind((host, port))
    server_socket.listen(5)

    print(f"Đang lắng nghe trên {host}:{port}...")

    while True:
        client_socket, addr = server_socket.accept()
        print(f"Kết nối từ: {addr[0]}:{addr[1]}")

        handle_client_connection(client_socket)
    
```

```python
main()
```
