```python
from preprocess import Preprocessor
import pandas as pd # type: ignore

import warnings

# Bỏ qua tất cả các cảnh báo
warnings.filterwarnings("ignore")

import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

```

```python
relative_path = "../../resources/data/101blockchains.csv"

df1 = pd.read_csv(relative_path)

df1.head()
```

```python
relative_path = "../../resources/data/101blockchains1.csv"

df2 = pd.read_csv(relative_path)

df2.head()
```

```python
relative_path = "../../resources/data/theblock.csv"

df3 = pd.read_csv(relative_path)

df3.head()
```

```python
relative_path = "../../resources/data/blockonomi0.csv"

df4 = pd.read_csv(relative_path)

df4.head()
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
df.head()
```

```python
df.info()
```

```python
preprocessor = Preprocessor()
```

```python
df[' Creation date'] = df[' Creation date'].apply(preprocessor.preprocess_date)
df.head()
```

```python
df.to_csv("../../resources/data/all_data.csv", index=False)
```

```python

df[' Article title'] = df[' Article title'].apply(preprocessor.preprocess_text)
df[' Author'] = df[' Author'].apply(preprocessor.preprocess_text)
df[' Content'] = df[' Content'].apply(preprocessor.preprocess_text)
df[' Category'] = df[' Category'].apply(preprocessor.preprocess_text)

df.head(10)
```

```python
dropped_columns = ['Article link', ' Website source', ' Article type', ' Summary', ' Temp']
df.drop(columns=dropped_columns, inplace=True)
df.head()
```

```python
df.to_csv("../../resources/data/preprocessed_data.csv", index=False)
```
